```python
import decaying


import numpy as np
import seaborn as sns

from matplotlib import pyplot as plt
```

## Linear

```python
linear = decaying.DecayingLinear(0.5, 1, 500)

X = range(1000)
y = [linear.step() for x in X]

sns.lineplot(x=X, y=y)
```

## Discount

```python
discount = decaying.DecayingDiscount(0.8, 1, 500)

X = range(1000)
y = [discount.step() for x in X]

sns.lineplot(x=X, y=y)
```
