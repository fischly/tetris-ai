```python
def test_t_spin_triple():
    # prepare field
    f = Field()
    f.field_data[14,:5] = 1

    f.field_data[15,:4] = 1

    f.field_data[16,:4] = 1
    f.field_data[16,5:] = 1

    f.field_data[17,0:4] = 1
    f.field_data[17,6:] = 1

    f.field_data[18:, :] = 1
    f.field_data[18,4] = 0

    # prepare piece
    p = Piece(0, [14,4])

    # f._print_board((f.add_piece_to_board(p, 0)))

    # perform t-spin triple
    (rot_succ, rot_result) = f._rotate(p, 0, 1)
    
    assert rot_succ == True, 'rotation must be successful'
    assert rot_result == (16, 3, 1), 't-piece must land in the t-spin triple hole'
    
    # f._print_board((f.add_piece_to_board(p, 1, [16,3])))

    
test_t_spin_triple()
```

```python
def test_j_kick():
    # prepare field
    f = Field()
    f.field_data[18,4:6] = 1
    f.field_data[16,3] = 1
    f.field_data[16,6] = 1
    f.field_data[15,4] = 1
    f.field_data[13,6] = 1
    
    
    # prepare piece
    p = Piece(2, [13,4])
    
    # f._print_board((f.add_piece_to_board(p, 0)))

    # perform t-spin triple
    (rot_succ, rot_result) = f._rotate(p, 0, -1)
        
    # f._print_board((f.add_piece_to_board(p, 3, [15,4])))
    assert rot_succ == True, 'rotation must be successful'
    assert rot_result == (15, 4, 3), 'j-piece must be kicked'
    
    # f._print_board((f.add_piece_to_board(p, 1, [16,3])))
```
